Inventory Management
●
Calculate the Economic Order Quantity using EOQ = √(2DS / H) by inputting
annual demand (D), ordering cost (S), and annual holding cost per unit (H).
○
Demand = units/year
○
Order cost = $/order
●
○
Holding cost = $/year/unit
Note: have user input demand, order cost and holding cost.

````python
import math

D = float(input("Enter annual demand (units/year): "))
S = float(input("Enter ordering cost ($/order): "))
H = float(input("Enter holding cost ($/year/unit): "))

EOQ = math.sqrt((2 * D * S) / H)

print(f"Economic Order Quantity (EOQ): {EOQ:,.2f} units")
````

Starting with an empty list, design a program in Python that populates the list by
repeatedly asking the user for numerical inputs.
The process stops when the user gives 0.
Finally, the list should be displayed in descending order.

````python
numbers = []

num = float(input("Enter a number (0 to stop): "))

while num != 0:
    numbers.append(num)
    num = float(input("Enter a number (0 to stop): "))

numbers.sort(reverse=True)

print(numbers)
````

#9
Dynamic dictionary; N distinct key:value pairs (while loop/ if-elif-else
statements)
●
Populate an empty dictionary iteratively N times (choose your own value for N) to add
name:count pairs. The key must be given as input from the user on every iteration. Use
conditionals to test if the key has been used before. If not, set count to 1; otherwise,
increase count by 1

````python
counts = {}
N = int(input("How many inputs? "))

i = 0
while i < N:
    name = input("Enter name: ")
    if name in counts:
        counts[name] += 1
    else:
        counts[name] = 1
    i += 1

print(counts)
````

Investment (ROI) using the formula: ROI = (revenue-cost)/cost

○
Make sure the cost is greater than 0 (otherwise give an error).
○
Make sure both inputs are numeric types (int or float).
○
Return the ROI as a decimal number (e.g., 0.2 for 20%).

````python
def roi(cost, revenue):
    if not (isinstance(cost, (int, float)) and isinstance(revenue, (int, float))):
        return "Error: inputs must be numeric"
    if cost <= 0:
        return "Error: cost must be greater than 0"
    return (revenue - cost) / cost
````

Loan Payment Calculator
●
●
Write a Python function called loan_payment() that calculates the monthly payment
on a loan using the standard loan formula: Payment = (r*P) / (1-((1+r)^
-n)) where:
○
P = loan amount (principal)
○
r = monthly interest rate (annual rate / 12)
○
n = total number of monthly payments (years × 12)
Your function should:
○
Take principal, annual
_
rate, and years as parameters
○
Validate that all inputs are positive numbers
○
Return the monthly payment

````python
def loan_payment(principal, annual_rate, years):
    if not (isinstance(principal, (int, float)) and isinstance(annual_rate, (int, float)) and isinstance(years, (int, float))):
        return "Error: inputs must be numeric"
    if principal <= 0 or annual_rate <= 0 or years <= 0:
        return "Error: inputs must be positive"

    r = annual_rate / 100 / 12
    n = years * 12

    payment = (r * principal) / (1 - (1 + r) ** (-n))
    return payment
````

Quality Control Manager (exceptions)
● A quality control manager would like to find the average weight of a list of weights. Prior
experience indicates that the first two measures are atypical, and the practice is to
exclude these from the average calculation.
● Write a program that calculates the average of a list of numbers that given as inputs from
the user, excluding the first three values.
● For your program, you must define and use the following functions:
○ getValues(n): Prompt the user for n values, returns the list
○ calcAverage(alist): Calculates the average from the list
● Use try-except to avoid invalid hours inputs (e.g., negative values, strings)

````python
def getValues(n):
    values = []
    count = 0
    while count < n:
        try:
            val = float(input())
            if val < 0:
                continue
            values.append(val)
            count += 1
        except:
            continue
    return values

def calcAverage(alist):
    if len(alist) <= 3:
        return 0
    trimmed = alist[3:]
    return sum(trimmed) / len(trimmed)

n = int(input())
values = getValues(n)
average = calcAverage(values)
print(average)
````

○ Attributes: first/last names
○ Method display(): takes a Boolean input parameter; displays the attribute values
either formally (last, first) or informally (first)
○ Special method __str__()

````python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def display(self, formal):
        if formal:
            print(f"{self.last}, {self.first}")
        else:
            print(self.first)

    def __str__(self):
        return f"{self.first} {self.last}"
````

Student class
● Create a class Student that has:
○ Three attributes: name (str), number (int), courses (list)
○ Two getter methods for name and courses
○ A setter method for name
○ The special method __len__ that returns the length of list courses
○ A method enroll() that appends a course to courses list if it is not there.
This is basically a more elaborate setter method.

````python
class Student:
    def __init__(self, name, number):
        self.name = name
        self.number = number
        self.courses = []

    def get_name(self):
        return self.name

    def get_courses(self):
        return self.courses

    def set_name(self, new_name):
        self.name = new_name

    def __len__(self):
        return len(self.courses)

    def enroll(self, course):
        if course not in self.courses:
            self.courses.append(course)
````

# 18
Class Employee - Manager Assignment

````python
class Employee:
    def __init__(self, name, hourly_wage, hours_worked):
        self.name = name
        self.__wage = 0
        self.set_wage(hourly_wage)
        self.hours_worked = hours_worked

    def get_wage(self):
        return self.__wage

    def set_wage(self, wage):
        if wage >= 0:
            self.__wage = wage
        else:
            print("Negative wages are not allowed.")

    def calculate_pay(self):
        return self.__wage * self.hours_worked

    def __str__(self):
        return f"Employee: {self.name}, Hourly Wage: ${self.__wage:.2f}, Hours Worked: {self.hours_worked}"


class Manager(Employee):
    def __init__(self, name, hourly_wage, hours_worked, bonus):
        super().__init__(name, hourly_wage, hours_worked)
        self.bonus = bonus

    def calculate_pay(self):
        return super().calculate_pay() + self.bonus

    def __str__(self):
        return f"Manager: {self.name}, Hourly Wage: ${self.get_wage():.2f}, Hours Worked: {self.hours_worked}, Bonus: ${self.bonus:.2f}"


employee = Employee("Heloisa", 20, 35)
manager = Manager("Sofia", 30, 40, 250)

print(employee)
print(f"Weekly Pay: ${employee.calculate_pay():.2f}")

print(manager)
print(f"Weekly Pay: ${manager.calculate_pay():.2f}")
````
