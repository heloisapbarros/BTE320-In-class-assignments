```python
def roi(cost, revenue):
    assert isinstance(cost, (int, float))
    assert isinstance(revenue, (int, float))
    assert cost != 0
    return (revenue - cost) / cost


def loan_payment(principal, annual_rate, months):
    assert isinstance(principal, (int, float))
    assert isinstance(annual_rate, (int, float))
    assert isinstance(months, (int, float))

    if principal <= 0:
        raise ValueError("Principal must be > 0")
    if months <= 0:
        raise ValueError("Months must be > 0")
    if annual_rate < 0:
        raise ValueError("Annual rate cannot be negative")

    r = annual_rate / 12

    if r != 0:
        return r * principal / (1 - (1 + r) ** (-months))
    else:
        return principal / months


try:
    cost = float(input("Enter cost: "))
    revenue = float(input("Enter revenue: "))
    print(roi(cost, revenue))

    principal = float(input("Enter principal: "))
    rate = float(input("Enter annual rate: "))
    months = float(input("Enter months: "))
    print(loan_payment(principal, rate, months))

except AssertionError:
    print("Programmer error")

except ValueError as e:
    print(e)

except Exception:
    print("Other error")
```
